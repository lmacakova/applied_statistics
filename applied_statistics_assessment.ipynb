### Task 1

```python
n=12
k=6
math.comb(n,k)
```

```python
labels = list(range(n))
labels
```

```python
combs = list(itertools.combinations(labels, k))


combs
```

```python
labels_n = random.sample(labels, 6)
labels_n
```

```python
set(labels_n)
```

```python
no_overlaps = []

for comb in combs:
  # Turn comb into a set.
  s1 = set(comb)
  # Turn labels_milk into a set.
  s2 = set(labels_n)
  # Figure out where they overlap.
  overlap = s1.intersection(s2)
  # Show the combination and the overlap.
  print(comb, overlap, len(overlap))
  # Append overlap to no_overlaps.
  no_overlaps.append(len(overlap))
```

```python
# Count the number of times each overlap occurs.
counts = np.unique(no_overlaps, return_counts=True)

# Show.
counts
```

```python
# Create a figure.
fig, ax = plt.subplots(figsize=(12, 8))

# Bar chart.
ax.bar(counts[0], counts[1])
```

##### What is normal distribution

A normal distribution or Gaussian distribution is a type of continuous probability distribution for a real-valued random variable. The general form of its probability density function is

$$ f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $$

The parameter  is the mean or expectation of the distribution (and also its median and mode), while the parameter 
$^2$ is the variance. The standard deviation of the distribution is . A random variable with a Gaussian distribution is said to be normally distributed, and is called a normal deviate.

##### Testing data

I am going to use 3 methods of testing normal distribution of data.
A histogram, a q-q plot and Shapiro-Wilk test

Histogram

```python

fig, ax = plt.subplots(figsize=(12, 8))
ax.hist(sample, bins=100, color='green', edgecolor='black')
```

A histogram is a plot showing frequency of values. Typical bell curve indicates normal distribution.

Shapiro-Wilk test

The Shapiro-Wilk test involves comparing the observed data with a corresponding set of normally distributed values. The null hypothesis of this test is that data follows normal distribution. The W statistic is derived from the correlation between the data and the corresponding normal scores. A value of W close to 1 suggests that the data follow a normal distribution. The p-value helps to determine whether to reject the null hypothesis of normality. A common threshold is 0.05. The p-value that is less than 0.05  indicates that data is not normally distributed

https://medium.com/@maizi5469/10-0-shapiro-wilk-test-5be38fd3c2a6

```python
stats.shapiro(sample)
```

I got W statistic close to 1 that indicates that data follows normal distribution. P-value is higher than 0.05 that could support null hypotesis, but I got warning that p-value could not be accurate for sample bigger than 5000 values. 

```python
stats.shapiro(sample[0:4999])
```

Shapiro-Wilk test with only first 5000 values. I can assume that data is normally distributed

Q - Q plot

```python

fig, ax = plt.subplots()
stats.probplot(sample, dist='norm', plot=ax)
```

A QQ plot (quantilequantile plot) is a probability plot, a graphical method for comparing two probability distributions by plotting their quantiles against each other.
https://en.wikipedia.org/wiki/Q%E2%80%93Q_plot
In this case, I compare my sample against hypotetical sample with normal distribution.As points fall onto 45 degree referrence line, there is strong indice that my sample has the normal distribution.
